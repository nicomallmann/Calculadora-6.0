```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


dataframe = pd.read_csv('/Gorjetas.csv')

sexos = ['Mulher','Homem']

for sexo in sexos:
  subset = dataframe[dataframe['sexo'] == sexo]

  sns.distplot(subset['gorjeta'], hist=False, kde=True,
               kde_kws={'linewidth' :3},
               label=sexo)
  

plt.legend(prop={'size':6}, title = 'Sexo')              
plt.title('Grafico de densidade de gorjeta por Sexo')
plt.xlabel('gorjeta')
plt.ylabel('Densidade')
plt.show()

  
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


dataframe = pd.read_csv('/Gorjetas.csv')

quantidades = [0,1,2,3,4,5,6]

for quantidade in quantidades:
  subset = dataframe[dataframe['quantidade'] == quantidade]

  sns.distplot(subset['gorjeta'], hist=False, kde=True,
               kde_kws={'linewidth' :3},
               label=quantidade)
  

plt.legend(prop={'size':6}, title = 'Quantidade')              
plt.title('Grafico de densidade de gorjetas por pessoas na mesa')
plt.xlabel('gorjeta')
plt.ylabel('Densidade')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


dataframe = pd.read_csv('/Gorjetas.csv')
dias = ['Qui', 'Sex', 'Sab', 'Dom']

for dia in dias:
  subset = dataframe[dataframe['dia'] == dia]

  sns.distplot(subset['gorjeta'], hist=False, kde=True,
               kde_kws={'linewidth' :3},
               label=dia)
  

plt.legend(prop={'size':6}, title = 'Dia da semana')              
plt.title('Gráfico de densidade de gorjetas por dias da semana')
plt.xlabel('gorjeta')
plt.ylabel('Densidade')
plt.show()
```
